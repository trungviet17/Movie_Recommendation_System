```python
import pandas as pd 
import numpy as np 
```

```python
from tmdbv3api import TMDb, Movie, Search
from os import getenv
tmdb = TMDb()

tmdb.api_key = getenv('TMDB_API_KEY')
search = Search()
```

```python
#  get genre 
movie = Movie()

def get_genre(id: int): 
    m = movie.details(id)
    genres = m.genres
    s = ''
    for j in genres : 
        s = s + ' ' + j['name']
    return s
```

```python
def get_score(id: int) : 
    m = movie.details(id)
    return m.vote_average
```

```python
def get_popularity(id: int) : 
    m = movie.details(id)

    return m.popularity 
```

```python
def get_overview(id: int): 
    m = movie.details(id)
    return m.overview 
```

```python
def get_or_lan(id: int): 
    m = movie.details(id)
    return m.original_language
```

```python
def get_casts(id: int): 
    m = movie.details(id)
    t = m.casts._json
    return dict(t)
```

```python

def separate(idx: int): 
    main_dataset = pd.DataFrame(columns=['Released','English title','TMDB_id'])

    for i in range(idx, idx + 4): 
        path = 'Data/' + str(i) + '.csv'
        tmp = pd.read_csv(path)
        tmp.drop(['index'], inplace= True, axis = 1)

        main_dataset = pd.concat([main_dataset,  tmp])

    main_dataset['Genre'] =  main_dataset['TMDB_id'].apply(lambda x: get_genre(x)) 
    main_dataset['Score'] = main_dataset['TMDB_id'].apply(lambda x: get_score(x))
    main_dataset['Popularity'] = main_dataset['TMDB_id'].apply(lambda x: get_popularity(x))
    main_dataset['Overview'] = main_dataset['TMDB_id'].apply(lambda x: get_overview(x))
    main_dataset['Or_lan'] = main_dataset['TMDB_id'].apply(lambda x: get_or_lan(x))
    main_dataset['Casts'] = main_dataset['TMDB_id'].apply(lambda x: get_casts(x))

    return  main_dataset
```
