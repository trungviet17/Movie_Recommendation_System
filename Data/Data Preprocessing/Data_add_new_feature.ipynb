```python
def get_score( m) : 
   
    return m.vote_average
```

```python
def get_popularity(m) : 
    
    return m.popularity 
```

```python
def get_overview(m): 
    
    return m.overview 
```

```python
def get_or_lan(m): 
    
    return m.original_language
```

```python
def get_casts(m): 
    
    t = m.casts._json
    t = dict(t)
    cast_list = []
    sort_t = sorted(t['cast'], key = lambda x: x['popularity'], reverse= True)
    for i in range(min(5, len(sort_t))) : 
        cast_list.append(sort_t[i]['id'])

    

    return cast_list
```

```python
def get_release(m): 
    return m.release_date
```

```python
def get_keyword(m): 

    k = m.keywords['keywords']
    list_keyword = []

    for i in k : 
        list_keyword.append(i['name'])

    return list_keyword
```

```python
def gather_information(title : str, year: int): 
    path = '..\Raw_data\\' + title + str(year) + '.csv' 

    df = pd.read_csv(path)
    series = pd.Series()
    genre = []
    score = []
    popularity = []
    overview = []
    language = []
    casts = []
    released = []
    keywords = []
    for i in range(len(df)) : 

        raw_data = movie.details(df.loc[i, 'TMDB_id'])
        
        cast_id = get_casts(raw_data)
        genre.append(get_genre(raw_data))
        score.append(get_score(raw_data))
        popularity.append(get_popularity(raw_data))
        overview.append(get_overview(raw_data))
        language.append(get_or_lan(raw_data))
        casts.append(cast_id)
        released.append(get_release(raw_data))
        keywords.append(get_keyword(raw_data))

        series = pd.concat([series, pd.Series(cast_id)])

    df['genre'] = pd.Series(data = genre)
    df['score'] = pd.Series(data=score)
    df['popularity'] = pd.Series(data=popularity)
    df['overview'] = pd.Series(data=overview)
    df['language'] = pd.Series(data = language)
    df['released'] = pd.Series(data =released )
    df['casts'] = pd.Series(data = casts)
    df['keyword'] = pd.Series(data = keywords)

    df.to_csv('..\Main_data\\' + title + str(year) + '.csv')
    series.to_csv('..\Main_data\\' + title + '_actor_id' + str(year) + '.csv')




    
```

## Add new user information 

```python
def get_bio (p): 
    return p.biography 

def get_birthday(p): 
    return p.birthday

def get_name(p): 
    return p.name 

def get_image(p): 
    return p.profile_path 

```

```python
person = Person()

def get_actor_info(title: str, year: int): 
    path = '..\Main_data\\' + title + '_actor_id' + str(year) + '.csv'

    df = pd.read_csv(path)
    df.drop_duplicates(inplace = True)
    df.dropna(inplace = True, ignore_index=True)
   
    
    name_list = []
    image_list = []
    birthday_list = []
    bio_list = []


    for i in range(len(df)): 
        print(df.loc[i, '0'])
        p = person.details(df.loc[i, '0'])
        
        
        bio_list.append(get_bio(p)) 
        name_list.append(get_name(p))
        image_list.append(get_image(p))
        birthday_list.append(get_birthday(p))

    main_df = pd.DataFrame(data = {
        'id' : list(df['0']), 
        'name'  : name_list, 
        'birthday' : birthday_list, 
        'bio'  : bio_list , 
        'image' : image_list
    })
    main_df.to_csv('..\Main_data\\' + title + '_actor_id' + str(year) + 'done.csv')

    return main_df
```

```python
for i in range(2010, 2012): 
    get_actor_info('American', i)
```
