```python
import pandas as pd 
import numpy as np 
import tmdbv3api
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_South_Korean_films_of_2023'

df_korean = pd.read_html(link, header= 0)

df_korean1 = df_korean[2]
df_korean2 = df_korean[3]
df_korean3 = df_korean[4]
df_korean4 = df_korean[5]

full_df  = pd.concat([df_korean1, df_korean2, df_korean3, df_korean4])
```

```python
# film from 2020 - 2023
full_df = pd.DataFrame(columns= ['English title','Native title'	,'Director(s)',	'Cast'	,'Ref'])

dct = {
    'january' : 1, 
    'february' : 2,
    'march' : 3, 
    'april' :  4,
    'may' : 5, 
    'june' : 6, 
    'july' : 7, 
    'august' : 8, 
    'september' : 9, 
    'october' : 10, 
    'november' : 11, 
    'december' : 12
}




for i in range(2020, 2024) : 
    link = 'https://en.wikipedia.org/wiki/List_of_South_Korean_films_of_'  + str(i)
    tmp_df = pd.read_html(link,  header= 0)

    full_tmp_df = pd.concat(tmp_df[2: 6])
    
    full_tmp_df.dropna(inplace= True)
    full_tmp_df['Opening'] = full_tmp_df['Opening'].str.lower()
    full_tmp_df['Opening'] = full_tmp_df['Opening'].str.replace(' ', '')
    full_tmp_df['Opening'] = full_tmp_df['Opening'].apply(lambda x : dct[x])
    full_tmp_df['Opening.1'] = full_tmp_df['Opening.1'].astype(int)
    full_tmp_df['Opening.1'] = full_tmp_df['Opening.1'].astype(str)
    full_tmp_df['Opening_day'] =  full_tmp_df['Opening'].astype(str) + '-' + full_tmp_df['Opening.1'] + '-' + str(i)
    full_tmp_df.drop(['Opening', 'Opening.1'], inplace = True, axis = 1)
    # full_tmp_df['Opening_day'] =  pd.to_datetime(full_tmp_df['Opening_day'])
    
    full_df = pd.concat([full_df, full_tmp_df])
    
```

```python
full_df.drop(['Ref', 'Ref.'], inplace = True, axis = 1)
full_df['Opening_day'] = pd.to_datetime(full_df['Opening_day'])
```

```python
search = Search()
def get_id(title : str, release_date) :
    t = release_date.year
    movie = search.movies(title, release_year=t)

   
    for j in movie['results'] : 
        
        if j['release_date'] == release_date.strftime('%Y-%m-%d') : 
            return j.id 
    return None 
```

```python
movie = Movie()

def get_genre(id : int ): 
    m = movie.details(id)
    genres = m.genres
    s = ''
    for j in genres : 
        s = s + ' ' + j['name']
   
    return s

full_df['Genres']= full_df['TMDB_ID'].apply(lambda x : get_genre(x))
```

```python
full_df.head()
```

```python
def get_score(id: int) : 
    m = movie.details(id)
    return m.vote_average

full_df['Score'] = full_df['TMDB_ID'].apply(lambda x : get_score(x))
full_df.head()
```

```python
def get_popularity(id: int) : 
    m = movie.details(id)

    return m.popularity 

full_df['Popularity'] = full_df['TMDB_ID'].apply(lambda x : get_popularity(x))
```

```python
full_df
```

```python
def get_overview(id: int): 
    m = movie.details(id)
    return m.overview 

full_df['Overview'] = full_df['TMDB_ID'].apply(lambda x : get_overview(x))
```

```python
def get_or_lan(id: int): 
    m = movie.details(id)
    return m.original_language
full_df['Original_Language'] = full_df['TMDB_ID'].apply(lambda x : get_or_lan(x))
```

```python
full_df.reset_index(drop= True)
```

```python
def get_casts(id: int): 
    m = movie.details(id)
    t = m.casts._json
    return dict(t)

full_df['Casts'] = full_df['TMDB_ID'].apply(lambda x : get_casts(x))
```
